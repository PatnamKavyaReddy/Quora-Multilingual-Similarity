```python
import pandas as pd
from googletrans import Translator
```

```python
df = pd.read_csv('train_translated.csv')
```

```python
df
```

```python
#Define object instance inside function so each call makes an instance
#If we use only one object instance, timeout will occur much faster due to too many requests being
#handled at one time
def translate2english(text):
    translator = Translator() 
    return translator.translate(text, dest = "en").text
```

```python
# need_translation_list_premise = []
# need_translation_list_hypothesis = []
# for index, row in df.iterrows():
#     if(row['lang_abv'] != 'en'):
#         need_translation_list_premise.append(row['premise'])
#         need_translation_list_hypothesis.append(row['hypothesis'])
```

```python
df[df['lang_abv'] == 'en'] #check how many rows are english so far
```

```python
#problem with googletrans is that you cannot make too many requests during a time. Thus we will keep translating
#in batches of 500 for rows that need translating (not english)
#so, run this and the next cell continuously until next cell shows that all premise/hypothesis has been translated!

i = 0
for index, row in df.iterrows():
    if(i != 300):
    
        if(row['lang_abv'] != 'en'):
            df.loc[index, 'premise'] = translate2english(row['premise'])
            df.loc[index, 'hypothesis'] = translate2english(row['hypothesis'])
            df.loc[index, 'lang_abv'] = 'en'
            i = i + 1
```

```python
df[df['lang_abv'] == 'en']
```

```python
#periodically save what has been translated so far into csv so you don't lose what has already been translated
df.to_csv("train_translated.csv", index = False)
```
